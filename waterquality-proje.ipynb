```python
for i in dataset:
  for j,k in enumerate(dataset[i].values):
    if k == '#NUM!':
      print(i,j)
```

```python
for column in dataset.columns:
    num_count = dataset[column].apply(lambda x: str(x).strip() == "#NUM" or str(x).strip() == "#NUM!").sum()  # '#NUM' değerlerini say
    if num_count > 0:
        print(f"'{column}' sütununda {num_count} adet '#NUM' değeri bulunmaktadır.")
```

```python
cleaned_dataset = dataset[dataset.iloc[:,:] != "#NUM!"]
```

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.DataFrame(cleaned_dataset)

scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df)

normalized_df = pd.DataFrame(normalized_data, columns=df.columns)

print("Normalleştirilmiş Veri Seti:")
print(normalized_df)
```

```python
x = normalized_df.iloc[:,:-1]
y = normalized_df.iloc[:,-1]
```

```python
x = x.dropna()
y = y[x.index]
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2,
                                                      random_state = 25)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
y_pred = rfc.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
print(cm,"\n")
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
rfcacc = accuracy_score(y_test,y_pred)
```

```python
import xgboost as xgb
xgb = xgb.XGBClassifier(objective="multi:softmax", num_class=3, random_state=42)
xgb.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
y_pred = xgb.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
print(cm,"\n")
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
xgbacc = accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(n_estimators=1000, learning_rate=0.1, max_depth=3, random_state=42)
gbc.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
y_pred = gbc.predict(x_test)
cm = confusion_matrix(y_test,y_pred)
print(cm,"\n")
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
gbcacc = accuracy_score(y_test,y_pred)
```
