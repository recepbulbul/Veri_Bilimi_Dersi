```python
for i in dataset:
  for j,k in enumerate(dataset[i].values):
    if k == '#NUM!':
      print(i,j)
```

```python
for column in dataset.columns:
    num_count = dataset[column].apply(lambda x: str(x).strip() == "#NUM" or str(x).strip() == "#NUM!").sum()  # '#NUM' deÄŸerlerini say
    if num_count > 0:
        print(f"'{column}' sÃ¼tununda {num_count} adet '#NUM' deÄŸeri bulunmaktadÄ±r.")
```

```python
cleaned_dataset = dataset[dataset.iloc[:,:] != "#NUM!"]
```

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.DataFrame(cleaned_dataset)

scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df)

normalized_df = pd.DataFrame(normalized_data, columns=df.columns)

print("NormalleÅŸtirilmiÅŸ Veri Seti:")
print(normalized_df)
```

```python
normalized_df.head()
```

```python
correlation_matrix = normalized_df.corr()
print(correlation_matrix)
```

```python
pearson_korelasyon = normalized_df.corr(method='pearson')
pearson_korelasyon
```

```python
spearman_korelasyon = normalized_df.corr(method='spearman')
spearman_korelasyon
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(pearson_korelasyon, annot=True, fmt=".2f", linewidths=0.5)
plt.title('Ã–zellikler ArasÄ±ndaki Korelasyon Heatmap')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(spearman_korelasyon, annot=True, fmt=".2f", linewidths=0.5)
plt.title('Ã–zellikler ArasÄ±ndaki Korelasyon Heatmap')
plt.show()
```
